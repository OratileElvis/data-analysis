```python
import pandas as pd

df1 = pd.read_csv("../datasets/WB_FINDEX_FIN26A.csv")
df2 = pd.read_csv("../datasets/WB_FINDEX_MOBILEACCOUNT_T_D.csv")

print("Dataset 1 preview:")
display(df1.head())

print("Dataset 2 preview:")
display(df2.head())
```

```python
import pandas as pd

df1_clean = df1[["REF_AREA", "REF_AREA_LABEL", "INDICATOR_LABEL",
                 "SEX", "AGE", "TIME_PERIOD", "OBS_VALUE"]]

df2_clean = df2[["REF_AREA", "REF_AREA_LABEL", "INDICATOR_LABEL",
                 "SEX", "AGE", "TIME_PERIOD", "OBS_VALUE"]]

df1_clean = df1_clean[df1_clean["REF_AREA"] == "ZAF"]
df2_clean = df2_clean[df2_clean["REF_AREA"] == "ZAF"]

print("Dataset 1 missing values:")
print(df1_clean.isnull().sum())

print("\nDataset 2 missing values:")
print(df2_clean.isnull().sum())

df1_clean = df1_clean.dropna(subset=["OBS_VALUE"])
df2_clean = df2_clean.dropna(subset=["OBS_VALUE"])

print("Dataset 1 shape after cleaning:", df1_clean.shape)
print("Dataset 2 shape after cleaning:", df2_clean.shape)

print("\nDataset 1 stats:")
display(df1_clean.describe())

print("\nDataset 2 stats:")
display(df2_clean.describe())

df1_clean.to_csv("../datasets/cleaned_dataset1.csv", index=False)
df2_clean.to_csv("../datasets/cleaned_dataset2.csv", index=False)
```
